```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time



def prothom_alo():
    driver.get("https://www.prothomalo.com")
    bsObj = BeautifulSoup(driver.page_source)

    prothom_alo_header = bsObj.find("div",{"class":"organism1-m__lead_story__1IQBR"}).find("a")
    print("Prothom Alo's Header News :",prothom_alo_header.attrs["aria-label"])
    print("Prothom Alo's Header News Link :",prothom_alo_header.attrs["href"])
    driver.get(prothom_alo_header.attrs["href"])
    
    time.sleep(3)
    
    total_height = int(driver.execute_script("return document.body.scrollHeight"))

    for i in range(1, total_height, 5):
        driver.execute_script("window.scrollTo(0, {});".format(i))
    
    print()


    
def jugantor():
    driver.get("https://www.jugantor.com")
    bsObj = BeautifulSoup(driver.page_source)

    jugantor_header = bsObj.find("div",{"id":"lead-news"})
    print("Jugantor's Header News :",jugantor_header.find("h3").get_text())
    print("Jugantor's Header News Link :",jugantor_header.find("a").attrs["href"])
    driver.get(jugantor_header.find("a").attrs["href"])
    
    time.sleep(3)
    
    total_height = int(driver.execute_script("return document.body.scrollHeight"))

    for i in range(1, total_height, 5):
        driver.execute_script("window.scrollTo(0, {});".format(i))
    
    print()

    
    
def bd_pratidin():
    driver.get("https://www.bd-pratidin.com")
    bsObj = BeautifulSoup(driver.page_source)

    bd_pratidin_header = bsObj.find("div",{"class":"news-tab-area latest-news-top"})
    print("Bangladesh Pratidin's Header News :",bd_pratidin_header.find("p").find("strong").get_text())
    print("Bangladesh Pratidin's Header News Link :","https://www.bd-pratidin.com/"+bd_pratidin_header.find("a").attrs["href"])
    driver.get("https://www.bd-pratidin.com/"+bd_pratidin_header.find("a").attrs["href"])
    
    time.sleep(3)
    
    total_height = int(driver.execute_script("return document.body.scrollHeight"))

    for i in range(1, total_height, 5):
        driver.execute_script("window.scrollTo(0, {});".format(i))
    
    print()


def kaler_kantho():
    driver.get("https://www.kalerkantho.com")
    bsObj = BeautifulSoup(driver.page_source)

    kaler_kantho_header = bsObj.find("div",{"class":"col-xs-6 leadnewsarea"})
    print("Kaler Kantho's Header News :",kaler_kantho_header.find("a",{"class":"n_head"}).get_text()[1:])
    print("Kaler Kantho's Header News Link :","https://www.kalerkantho.com"+kaler_kantho_header.find("a",{"class":"n_head"}).attrs["href"][1:])
    driver.get("https://www.kalerkantho.com"+kaler_kantho_header.find("a",{"class":"n_head"}).attrs["href"][1:])
    
    time.sleep(3)
    
    total_height = int(driver.execute_script("return document.body.scrollHeight"))

    for i in range(1, total_height, 5):
        driver.execute_script("window.scrollTo(0, {});".format(i))
    
    print()
    

def daily_star():
    driver.get("https://www.thedailystar.net")
    bsObj = BeautifulSoup(driver.page_source)

    daily_star_header = bsObj.find("div",{"class":"columns small-12 medium-12 large-4 lead-news"})
    print("Daily Star's Header News :",daily_star_header.find("h3").get_text())
    print("Daily Star's Header News Link :","https://www.thedailystar.net"+daily_star_header.find("h3").find("a").attrs["href"])
    driver.get("https://www.thedailystar.net"+daily_star_header.find("h3").find("a").attrs["href"])
    
    time.sleep(3)
    
    # this part is for scrolling down the pages
    total_height = int(driver.execute_script("return document.body.scrollHeight"))

    for i in range(1, total_height, 5):
        driver.execute_script("window.scrollTo(0, {});".format(i))
    # this part is for scrolling down the pages
    
    print()
    


def samakal():
    driver.get("https://samakal.com")
    bsObj = BeautifulSoup(driver.page_source)

    samakal_header = bsObj.find("div",{"class":"leadDesign2"})
    print("Samakal's Header News :",samakal_header.find("h3").get_text())
    print("Samakal's Header News Link :",samakal_header.find("h3").find("a").attrs["href"])
    driver.get(samakal_header.find("h3").find("a").attrs["href"])
    
    time.sleep(3)
    
     #_______________this part is for scrolling down the pages_____________
    
    total_height = int(driver.execute_script("return document.body.scrollHeight"))

    for i in range(1, total_height, 5):
        driver.execute_script("window.scrollTo(0, {});".format(i))
        
    #_______________this part is for scrolling down the pages_____________
    
   
    
    print()
    

    
def ittefaq():
    driver.get("https://www.ittefaq.com.bd")
    bsObj = BeautifulSoup(driver.page_source)

    ittefaq_header = bsObj.find("div",{"class":"row"})
    print("Ittefaq's Header News :",ittefaq_header.find("h2").get_text())
    print("Ittefaq's Header News Link :","https:"+ittefaq_header.find("h2").find("a").attrs["href"])
    driver.get("https:"+ittefaq_header.find("h2").find("a").attrs["href"])
    
    time.sleep(3)
    
    #_______________this part is for scrolling down the pages_____________
    
    total_height = int(driver.execute_script("return document.body.scrollHeight"))

    for i in range(1, total_height, 5):
        driver.execute_script("window.scrollTo(0, {});".format(i))
        
    #_______________this part is for scrolling down the pages_____________
    
    
    print()
    

path = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(path)


prothom_alo()

daily_star()

bd_pratidin()

samakal()

ittefaq()

kaler_kantho()

jugantor()


driver.quit()
```
