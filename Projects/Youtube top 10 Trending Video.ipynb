# Collecting information for first video

```python
from selenium import webdriver
from bs4 import BeautifulSoup

path = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get("https://www.youtube.com/feed/trending")

bsObj = BeautifulSoup(driver.page_source)

first_video = bsObj.find("ytd-video-renderer")
print(first_video.prettify())

print()

# title
first_video_title = first_video.find("yt-formatted-string",{"class":"style-scope ytd-video-renderer"})
print("Title :",first_video_title.get_text())

print()

# channel name
first_video_channel_name = first_video.find('ytd-channel-name')
print("Channel Name :",first_video_channel_name.find("a").get_text())

print()

# no of views
first_video_views = first_video.find("div",{"id":"metadata-line"}).findAll("span")[0]
print("No of Views :",first_video_views.get_text())

print()

# the date when it got uploaded
first_video_uploaded = first_video.find("div",{"id":"metadata-line"}).findAll("span")[1]
print("Date of Getting Uploaded :",first_video_uploaded.get_text())

print()

# description
first_video_description = first_video.find("yt-formatted-string",{"id":"description-text"})
print("Video Description :",first_video_description.get_text())

print()

# video url/link
first_video_url = first_video.find("a",{"id":"video-title","class":"yt-simple-endpoint style-scope ytd-video-renderer"})
print("Video url :",first_video_url.attrs["href"])

actual_first_video_url = "https://www.youtube.com"+first_video_url.attrs["href"]
print("Actual Video url :",actual_first_video_url)

print()

# thumbnail url/link
first_video_thumbnail_url = first_video.find("a",{"id":"thumbnail"}).find("img")
print("Thumbnail url :",first_video_thumbnail_url.attrs["src"])
```
