```python
from selenium import webdriver
from bs4 import BeautifulSoup

path = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get("https://www.shohoz.com/booking/bus/search?fromcity=Rajshahi&tocity=Dhaka&doj=24-Sep-2022&dor=")

bsObj = BeautifulSoup(driver.page_source)

#first_bus = bsObj.find("div",{"class":"container"}).find("div",{"class":"row"}).find("div",{"class":"col-md-9"}).find("div",{"id":"bus_tckt_rows"}).find("tbody").find("tr")
first_bus = bsObj.find("table",{"id":"trips"}).find("tbody").find("tr",{"class":"trip-row"})
#print(first)

bus_company = first_bus.findAll("td")[0].findAll("li")[0].get_text()
print("Bus Company :",bus_company)

bus_type = first_bus.findAll("td")[0].findAll("li")[1].get_text()
print("Bus Type :",bus_type)

route = first_bus.findAll("td")[0].findAll("li")[2].get_text()[7:]
print("Route :",route)

starting_point = first_bus.findAll("td")[0].findAll("li")[3].get_text()[17:]
print("Starting Point :",starting_point)

ending_point = first_bus.findAll("td")[0].findAll("li")[4].get_text()[15:]
print("Ending Point :",ending_point)

departure_time = first_bus.findAll("td")[1].get_text().strip()
print("Departure Time :",departure_time)

arrival_time = first_bus.findAll("td")[2].get_text().strip()
print("Arrival Time :",arrival_time)

available_seat = first_bus.findAll("td")[3].get_text().strip()
print("Available Seats :",available_seat)

fare = first_bus.findAll("td")[4].find("li").get_text().strip()
print("Fare :",fare)
```
