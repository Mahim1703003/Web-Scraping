```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "https://en.wikipedia.org/wiki/"
director_name = input("Enter a director's name : ")
director_name = director_name.replace(" ","_")


html = urlopen(url+director_name)
bsObj = BeautifulSoup(html.read())


directed_feature = bsObj.find("caption").find_next_sibling()

all_rows = directed_feature.findAll("tr")

count = -1
column_names = list()
years = list()
titles = list()
distributors = list()
movie_links = list()

for row in all_rows:
    if count==-1:
        for th_tag in row.findAll('th'):
            column_names.append(th_tag.get_text().strip())  
    else:
        l = row.findAll('td')

        if len(l)==3:
            years.append(l[0].get_text().strip())
            titles.append(l[1].get_text().strip())
            distributors.append(l[2].get_text().strip())

        if len(l)== 1:
            titles.append(l[0].get_text().strip())
            years.append(years[count-1])
            distributors.append(distributors[count-1])

        if len(l)==2:
            italic_list = l[0].findAll("i")
            if len(italic_list) == 0:
                years.append(l[0].get_text().strip())
                titles.append(l[1].get_text().strip())
                distributors.append(distributors[count-1])
            else:
                titles.append(l[0].get_text().strip())
                distributors.append(l[1].get_text().strip())
                years.append(years[count-1])
                
        movie_links.append(row.find("a").attrs['href'])
    
    
    count = count + 1



df = pd.DataFrame({
    column_names[0] : years,
    column_names[1] : titles,
    column_names[2] : distributors
},  columns = column_names)

df
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "https://en.wikipedia.org"

running_time = list()
country = list()
language = list()

for movie_link in movie_links:
    html = urlopen(url+movie_link)
    bsObj = BeautifulSoup(html.read())
    
    info_rows = bsObj.find("table",{"class":"infobox vevent"})
    
    if info_rows.find("th",string="Running time") == None:
        running_time.append("")
    else:
        running_time.append(info_rows.find("th",string="Running time").find_next_sibling().get_text())
            
    if info_rows.find("th",string=re.compile("^(Countr).*")) == None:
        country.append("")
    else:
        country.append(info_rows.find("th",string=re.compile("^(Countr).*")).find_next_sibling().get_text().strip())
            
    if info_rows.find("th",string="Language") == None:
        language.append("")
    else:
        language.append(info_rows.find("th",string="Language").find_next_sibling().get_text())

movie_df = pd.DataFrame({
    "Movie" : titles,
    "Running time" : running_time,
    "Country" : country,
    "Language" : language
}, columns = ['Movie','Running time','Country','Language'])

movie_df.set_index('Movie')
```
